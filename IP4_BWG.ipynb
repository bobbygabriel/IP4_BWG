## Question 3
Create a 1D array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X.

a) Create a plot using plt.subplots() and plot X versus X^2

```python
x = np.linspace(0, 100, 100)
y = np.square(x)

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set_xlabel('X')
ax.set_ylabel('X^2')
ax.set_title('X vs X^2')

plt.show()
```

b) create a scatter plot of x versus the exponential of x

```python
x = np.linspace(0, 100, 100)
y = np.exp(x)

fig, ax = plt.subplots()

ax.scatter(x, y)

ax.set_xlabel('X')
ax.set_ylabel('Sqrt(x)')
ax.set_title('X vs Sqrt(x)')

plt.show()
```

## Question 5
Remove all symbols from the price column and convert and update it to 64 bit int

```python
car_sales['Price'] = car_sales['Price'].str.replace('$', '')
car_sales['Price'] = car_sales['Price'].str.replace(',', '')
car_sales['Price'] = car_sales['Price'].astype('int64')

car_sales.head()
```

```python
car_sales.info()
```

## Question 6
Add a column called 'Total Sales' to car_sales which cumulatively adds the Price column.

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head()
```

## Question 7
Add a column called 'Sale Date' which lists a series of succesive dates starting from today. 

```python
import datetime as dt
```

```python
start_date = dt.datetime.today()
dates = pd.date_range(start=start_date, periods=len(car_sales), freq='D').date
car_sales['Sale Date'] = dates

car_sales.head()
```

## Question 16
Find out Pearson correlation between Total Slae, Price, Odometer. Comment on the correlation

```python
correlation = car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')

cols=['Total Sales', 'Price', 'Odometer (KM)']

sns.heatmap(correlation, annot=True, xticklabels=cols, yticklabels=cols)
```

Price and Total Sales have the highest correlation to one another at .16
Price and Odometer (KM) have the lowest correlation to one another at -0.63
